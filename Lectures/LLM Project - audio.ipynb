## Working wiith audio (CPU possible)

```python
from transformers import WhisperProcessor, WhisperForConditionalGeneration
```

```python
# The demo model is only for English
processor = WhisperProcessor.from_pretrained("openai/whisper-tiny.en")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-tiny.en")
```

Consider https://huggingface.co/openai/ for other variants

```python
audio_src = "https://traffic.libsyn.com/blogrelations/monkey-flea.mp3"
```

```python
# to fetch urls from the web
import requests

url = 'http://example.com/somefile.mp3'
response = requests.get(audio_src)

with open('output.mp3', 'wb') as file:
    file.write(response.content)
```

```python
from IPython.display import Audio
```

```python
# If you want to hear the story
Audio("output.mp3")
```

```python
import librosa

# Load the mp3 file
waveform, sample_rate = librosa.load('output.mp3', sr=16000)
```

```python
# audio signal converted to a mel spectrogram (cf. deep learning lectures)
input_features = processor(waveform, sampling_rate=sample_rate, return_tensors="pt")['input_features']
input_features.shape
```

```python
# Generate token ids
predicted_ids = model.generate(input_features)
```

```python
# Decode token ids to text
transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)
```

```python
from IPython.display import Markdown
Markdown(transcription[0])
```

You only have 30 second of story because that is how whisper works. You can go beyond this either by using command line instruction (not available here), or by splitting the file in chunks.

```python
!whisper output.mp3 # won't work
```

```python
import numpy as np

# Define chunk size in seconds
chunk_size_seconds = 30

# Calculate chunk size in data points
chunk_size = chunk_size_seconds * sample_rate

# Split the audio signal into chunks
chunks = np.split(waveform, np.arange(chunk_size, len(waveform), chunk_size))
```

```python
story = []
for chunk in chunks:
  input_features = processor(chunk, sampling_rate=sample_rate, return_tensors="pt")['input_features']
  predicted_ids = model.generate(input_features)
  transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)
  story.append(transcription[0])
```

```python
combined_story = " ".join(story)
Markdown(combined_story)
```
